```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
# Kepler Exoplanet Light Curves Time Series
# Source: http://keplerscience.arc.nasa.gov
# Light curve in relative flux versus phase
```

```python
from skdaccess.astro.kepler import DataFetcher as Kepler_DF
from skdaccess.framework.param_class import *
import numpy as np
```

```python
kepler_fetcher = Kepler_DF([AutoList(['009941662','10001893'])], filter_window=50/24, normalize=True)
```

```python
kepler_data = kepler_fetcher.output().get()
```

```python
kepler_data['009941662'].head()
```

```python
stellar_data = kepler_data['009941662']
```

```python
plt.figure(figsize=(12,6));
data = stellar_data.loc[stellar_data.index<437.38,'PDCSAP_FLUX'].dropna()
plt.plot(np.array(data.index) % 1.7636, data,'.');
plt.tight_layout();
```
